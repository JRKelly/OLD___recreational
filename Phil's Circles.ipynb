```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import lines
import math
from matplotlib.backends.backend_pdf import PdfPages 
```

```python
def draw_circles(R, RADIUS, NUMBER):
    plt.axes()
    bigcircle = plt.Circle((0, 0), radius=R, fc='none')
    plt.gca().add_patch(bigcircle)
    X0 = 0
    Y0 = R
    a = 2*np.pi/NUMBER
    angles = [a/2]
    for j in range (1,NUMBER): angles.append(a*j+a/2)
    color = 0
    for angle in angles:
    
        X = R*math.cos(angle)
        Y = R*math.sin(angle)
        
        color += 0.5
        #color1 = str(color)
       
        
        smallcircle = plt.Circle((X, Y), radius=RADIUS, fc = (0, color / 20.0, 0, 1))
        plt.gca().add_patch(smallcircle)
        plt.axis('scaled')
    
```

```python
k = 1
circum28 = 28*k
circum19 = 19*k
R28 = circum28/(2*np.pi)
R19 = circum19/(2*np.pi)
littleR = k/2.0
```

```python
plot1 = plt.figure(1)
draw_circles(R28, littleR, 28)
draw_circles(R19, littleR, 19)
earth = plt.Circle((0, 0), radius=littleR, fc='none')
plt.gca().add_patch(earth)
plt.axhline(y=0, xmin=0.5, xmax=1)
plt.show()
pp = PdfPages('figure2.pdf')
pp.savefig(plot1)
pp.close()
```
